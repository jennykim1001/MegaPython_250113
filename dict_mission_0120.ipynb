**[실습1] 친구 연락처 검색하는 프로그램을 작성해 봅시다.**
* addr 딕셔너리 : 6명의 친구이름과 연락처를 임의대로 저장
* 친구이름을 입력하면 친구의 연락처를 출력
* 무한 반복(while True) 하는데, 0 이 입력되면 “프로그램 종료합니다!” 메시지 출력하고 무한 반복을 빠져나옴(break)
* 등록되지 않은 친구이름을 입력하면 “등록되지 않은 친구입니다” 메시지 출력

```
[실행 결과]
검색가능한 친구들:  ['최재원', '유재석', '김연수', '김우정', '김가현', '김혜현']
친구이름 이름 입력(검색 종료는 0) : 유재석
유재석의 연락처 : 010-2222-1234
친구이름 이름 입력(검색 종료는 0) : 유인아
등록되지 않은 친구입니다.
친구이름 이름 입력(검색 종료는 0) : 0
프로그램을 종료합니다!
```




````python
addr={'최재원':'010-1111-1234', '유재석':'010-2222-1234', '김연수':'010-3333-1234',
      '김우정':'010-4444-1234', '김가현':'010-5555-1234', '김혜현':'010-6666-1234'}

# print('검색 가능한 친구들: ', list(addr.keys()))
print(f'검색 가능한 친구들 : {list(addr.keys())}')

while True:
  name = input('친구이름 이름 입력(검색 종료는 0) : ')

  if name in addr:
    print(f'{name}의 연락처 : {addr[name]}')

  elif name == '0':
    print('프로그램 종료합니다!!')
    break
  else:
    print('등록되지 않은 친구입니다')
````

**[실습2] 앞서 작성한 “친구 연락처 검색 프로그램”을 다음과 같이 만들어보세요.**
* 추가 : 딕서너리에 친구이름, 친구전호번호 추가
* 검색 : 친구이름 입력하면 전화번호 검색
* 0 입력하면 프로그램 종료

```
[실행결과]

▶ 메뉴 선택(1:추가, 2:검색, 0:종료)? 1
친구 이름 입력>> 홍길동
전화번호 입력>> 010-1234-0000

▶ 메뉴 선택(1:추가, 2:검색, 0:종료)? 2
검색가능한 친구들:  ['최재원', '유재석', '김연수', '김우정', '김가현', '김혜현', '홍길동']
친구이름 입력>> 유재석
유재석의 연락처 : 010-2222-1234

▶ 메뉴 선택(1:추가, 2:검색, 0:종료)? 5
메뉴 선택 오류입니다. 다시 선택하세요.

▶ 메뉴 선택(1:추가, 2:검색, 0:종료)? 0
프로그램을 종료합니다!
```



````python
addr={'최재원':'010-1111-1234', '유재석':'010-2222-1234', '김연수':'010-3333-1234',
      '김우정':'010-4444-1234', '김가현':'010-5555-1234', '김혜현':'010-6666-1234'}

while True:
  menu = int(input('▶ 메뉴 선택(1:추가, 2:검색, 0:종료)? '))

  #데이터 추가
  if menu == 1:
    addr_key = input('친구 이름 입력>> ')
    if addr_key not in addr:
      addr_value = input('전화번호 입력>> ')
      addr[addr_key] = addr_value
    else:
      print('존재하는 친구입니다.')

  #데이터 조회
  elif menu == 2:
    print(f'검색가능한 친구들 : {list(addr.keys())}')
    search_name = input('검색할 친구이름 입력>> ')
    if search_name in addr:
      print(f'{search_name}의 연락처 : {addr[search_name]}')

  elif menu == 0:
    print('프로그램을 종료합니다!')
    break
  else:
    print('메뉴 선택 오류입니다. 다시 선택해주세요')
    continue
````

````python
# 회원가입 프로그램
members={}

while True :
  menu = int(input('''
  1. 회원가입
  2. 프로그램 종료
  >> '''))

  if menu == 1:
    id = input('아이디 입력>> ')
    pw = input('비밀번호 입력>> ')
    if id in members:
      print('존재하는 아이디입니다')
      print('다시 가입해주세요')
    else:
      members[id]=pw
      print('등록완료')

  elif menu == 2:
    break
  else:
    print('입력 오류')

print('[회원 목록]')
print('아이디\t비밀번호')
print('-'*20)
for key, value in members.items():
  print(f'{key}\t{value}')
print('-'*20)

print('Bye~~~')
````
